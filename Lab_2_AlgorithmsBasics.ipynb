```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
st=time.time()
func2(2, 4)
et=time.time()
f=(et-st)*1000
print("the time required for the function is:",f*1000)
# this funtion return the 2^4
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
st=time.time()

print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
et=time.time()
f=(et-st)*1000
print("the time required for the function is:",f*1000,"ms")

#this function returns the common number of element count in both the given array
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
#this function returns the number of attempt required to find the number enter by the user
st=time.time()

print(func5(25))
et=time.time()
f=(et-st)*1000
print("the time required for the function is:",f*1000,"ms")
# Sample function call: func5(16)
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
st=time.time()

print(func12([1, 2, 3, 4, 5], 2))
et=time.time()
f=(et-st)*1000
print("the time required for the function is:",f*1000,"ms")
# this function returns a custom array from the old index with the formula of getting new index= (i+k)% length of given array
```

##1.2 Arrays
Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  a_len = len(arr)
  total_sum = 0

  for i in range(0, a_len):
       total_sum = total_sum + arr[i]
  return total_sum
```

```python

start=time.time()
arr = [12, 3, 4, 15,45,78,450]
print(sum_arr(arr))
end=time.time()
print("the time required for the function is:",(end-start)*1000,"ms")
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  a_len = len(arr)
  max = arr[0]
  min = arr[0]
  for i in range(0,a_len):
    if arr[i]>max:
      max=arr[i]
    elif arr[i]<min:
      min=arr[i]
  return max,min
```

```python
start=time.time()
arr = [12, 3, 4, 15,45,78,450]
print(max_min(arr))
end=time.time()
print("the time required for the function is:",(end-start)*1000,"ms")
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
  a_len = len(arr)
  ec=0
  oc=0
  for i in range(0,a_len):
    if(arr[i]%2==0):
      ec=ec+1
    else:
      oc=oc+1
  print("even number count=",ec)
  print("odd number count =",oc)

```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr,tar):
  l=0
  h=len(arr)-1
  while l<=h:
    mid = (l+h)//2
    if(arr[mid]==tar):
      return mid
    elif arr[mid]<tar:
      l=mid+1
    else:
      h=mid-1
  return -1




```
